```python
dd = {'Year': data['Year'],
       'PST': data['PST'],
       'Global Horizontal [W/m^2]': data['Global Horizontal [W/m^2]'],
      
       'Diffuse Horizontal [W/m^2]': data['Diffuse Horizontal [W/m^2]'],
       'Air Temperature [deg C]': data['Air Temperature [deg C]'],
      'Avg Wind Speed @ 3m [m/s]': data['Avg Wind Speed @ 3m [m/s]'],
      'Rel Humidity [%]': data['Rel Humidity [%]'],
       'Direct Normal [W/m^2]': data['Direct Normal [W/m^2]']
     }
```

```python
df = pd.DataFrame(data = dd)
```

```python
df
```

```python
X= df.iloc[:, 2:7]
print(X)
```

```python
Y= df.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25)
```

```python
regr = LinearRegression(normalize = True)
regr.fit(X_train, Y_train)
regr.fit(X_test, Y_test)
```

```python
regr.score(X_train, Y_train)
```

```python
regr.score(X_test, Y_test)
```

```python
from sklearn import model_selection
```

```python
print(model_selection.cross_val_score(regr, X_test,Y_test))
```

```python
y_pred = regr.predict(X_test)
```

```python
diff = abs(y_pred[0:-1] - Y_test[0:-1])
mae = diff.mean()
print(mae)
```

```python
diff2 = diff**2
rmse = np.sqrt(diff2.mean())
print(rmse)
```

```python
print(regr.coef_)
```

```python
print(regr.intercept_)
```
